```python
closing = data['Close'] 
closing.head()
```

```python
from pandas import Series 
cdate = data['Date'].values
cdate = cdate[1:] # this will return the date for your index 
N = closing.size
change = closing[:N-1].values-closing[1:].values # thiscalculates the price di erence
changeData = Series(change, index=cdate) # this creates a new Series
changeData.head() # display the  rst 5 rows of the Series
```

```python
import matplotlib
%matplotlib inline
changeData.hist(normed=True)
```

```python
changeData.mean()
```

```python
changeData.std()
```

```python
Z = (changeData-changeData.mean())/changeData.std()
```
