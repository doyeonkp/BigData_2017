```python
import json
f = open('wh.json')
tweets = [json.loads(line) for line in f]
f.close()
```

```python
from collections import defaultdict

counts = defaultdict(int)
for tweet in tweets:
    words = tweet['text'].split(' ')
    for word in words:
        counts[word] += 1
```

```python
import operator
```

```python
sorted_words = sorted(counts.items(), key=operator.itemgetter(1))
sorted_words[-15:]
```
